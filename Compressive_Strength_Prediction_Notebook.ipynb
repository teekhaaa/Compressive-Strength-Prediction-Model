### Observation
-- This shows that there is outliers in Slag, Superplasticers and age column

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Detect outliers
outliers_iqr = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("Number of outliers detected per feature using IQR:", outliers_iqr)
```

```python
upper_limit = df > Q3 + 1.5 * IQR
lower_limit = df < Q1 - 1.5 * IQR
```

```python
upper_limit.sum()
```

```python
lower_limit.sum()
```

```python
Q1 = df[['slag', 'superplasticizer', 'age']].quantile(0.25)
Q3 = df[['slag', 'superplasticizer', 'age']].quantile(0.75)
IQR = Q3 - Q1

# Calculate the upper and lower bounds for outliers
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
```

```python
# Filter rows where values in 'slag', 'superplasticizer', 'age' are within the limits
df = df[
    (df['slag'] >= lower_limit['slag']) & (df['slag'] <= upper_limit['slag']) &
    (df['superplasticizer'] >= lower_limit['superplasticizer']) & (df['superplasticizer'] <= upper_limit['superplasticizer']) &
    (df['age'] >= lower_limit['age']) & (df['age'] <= upper_limit['age'])
]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('csMPa', axis = 1)
y = df['csMPa']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_pred = lr.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
df['csMPa'].describe()
```

- Since MSE is 56.61, it seems relatively high compared to the scale of your target values (mean of 35 and max of 82).
- We have to improve our model, as the current error suggests that predictions deviate significantly from the actual values.
- Lowering the MSE closer to the mean would generally indicate better performance.

```python
r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)
```

- An R² score of 0.79 means that linear regression model explains about 79% of the variance in the target variable (csMPa).
- However, there is still around 21% of the variance that the model is not explaining

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cv_scores = cross_val_score(lr, x, y, cv=5)
print("Cross-Validation Scores:", cv_scores) 
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)  # Try different degrees (2, 3)
x_poly = poly.fit_transform(x)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred_rf)
print("R² Score by RandomForest:", r2)
```

```python
mse_rf = mean_squared_error(y_test, y_pred_rf)
print("Random Forest Mean Squared Error:", mse_rf)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10] 
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5)
grid_search.fit(x_train, y_train)

print("Best Parameters:", grid_search.best_params_)   
```

```python
importances = rf.feature_importances_
sorted_importances = sorted(zip(importances, x.columns), reverse=True)
print("Feature Importance:", sorted_importances)  
```

```python
best_rf = RandomForestRegressor(max_depth=None, min_samples_split=2, n_estimators=200, random_state=42)
best_rf.fit(x_train, y_train)
y_pred_best = best_rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print("Best Model MSE:", mse_best)
print("Best Model R² Score:", r2_best)
```

```python
import joblib
joblib.dump(best_rf, 'concrete_strength_model.pkl')
```
